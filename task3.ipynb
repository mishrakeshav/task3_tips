```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset("tips") 
```

```python
ax=sns.violinplot(x="total_bill",data=tips)
plt.show()
```

```python
fig,ax=plt.subplots(0)
tips.plot(x="total_bill",y="tip",kind="scatter")
```

```python
tips.plot("day","tip",width=0.2,kind="hist")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
ax=sns.catplot(x="time", y="total_bill", hue="smoker", data=tips)
plt.show()
```

```python
ax=sns.countplot(x="time", hue="smoker", data=tips)
plt.show()
```
